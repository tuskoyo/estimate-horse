```python
import pandas as pd
import numpy as np
import os
df = pd.read_csv('/home/osamu/horse_data/c100_esti/position_id.csv',index_col = 0)
df.info()
```

```python
#各行において欠損値が出る場合は、その行を削除する
df = df.dropna()
#各列において一つでも負の値が出る場合は、その行を削除する
df = df[(df > 0).all(axis = 1)]
```

```python
df
```

```python
ori_dir = '/home/osamu/horse_data/c100_esti/'
new_dir = '/home/osamu/horse_data/c100_crean_data/'
```

```python
os.makedirs(new_dir,exist_ok = True)
#ori_dirからdf.indexにあるファイルをnew_dirにコピーする
for file in df.index:
    file_path = os.path.join(ori_dir,f'{file}.jpg')
    new_path = os.path.join(new_dir,f'{file}.jpg')
    if os.path.exists(file_path):
        os.system('cp {} {}'.format(file_path,new_path))
    else:
        print('file not found:',file_path)
```

```python
df.to_csv(os.path.join(new_dir,'position_id.csv'))
```
